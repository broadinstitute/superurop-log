Load FCOS model from torchvision

```python
import torchvision
import torch
import cv2
```

```python
model = torchvision.models.detection.fcos_resnet50_fpn(pretrained=False)
model.eval()
```

Example inference run with image from the Kaggle Broad Data Science bowl

```python
img = cv2.imread("BBBC038_sample_img.png")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = img / 255.0
img = torch.from_numpy(img)
img = img.float()
img = img.permute(2, 0, 1)
img = img.unsqueeze(0)
```

```python
out = model(img)
```
